```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")


X_test = test_df
```

```python
#data preprocessing
```

```python
X_train = train_df.drop('Survived' , axis=1)
Y_train = train_df['Survived']
```

```python
test_df.head()
```

```python
X_train.head()
```

```python
X_train.head()

X_train['Title'] = X_train['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
X_test['Title'] = X_test['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
X_test.head()
```

```python
gender_mapping = {'male': 0 , 'female': 1}
gender_map = {'Mr' : 0 , 'Master' : 0 , 'Mrs' : 1 , 'Miss' : 1}
```

```python
X_train['Sex'] = X_train['Sex'].map(gender_mapping)
X_test['Sex'] = X_test['Sex'].map(gender_mapping)
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train['Fare'] = scaler.fit_transform(X_train[['Fare']])
X_test['Fare'] = scaler.fit_transform(X_test[['Fare']])
```

```python
X_train.head()
```

```python

X_train.drop('Name' , axis = 1 , inplace= True)
X_test.drop('Name' , axis = 1 , inplace= True)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train['Age'].fillna(X_train['Age'].median() , inplace = True)
X_test['Age'].fillna(X_test['Age'].median() , inplace = True)

Y_train.head()

X_train['Title'] = X_train['Title'].map(gender_map)
X_test['Title'] = X_test['Title'].map(gender_map)
```

```python
most_frequent_title = X_train['Title'].mode().iloc[0]
X_train['Title'].fillna(most_frequent_title, inplace=True)

most_frequent_title = X_test['Title'].mode().iloc[0]
X_test['Title'].fillna(most_frequent_title, inplace=True)

most_frequent_TICKET = X_train['Ticket'].mode().iloc[0]
X_train['Ticket'].fillna(most_frequent_title, inplace=True)

most_frequent_title = X_test['Ticket'].mode().iloc[0]
X_test['Ticket'].fillna(most_frequent_title, inplace=True)

most_frequent_title = X_train['Cabin'].mode().iloc[0]
X_train['Cabin'].fillna(most_frequent_title, inplace=True)

most_frequent_title = X_test['Cabin'].mode().iloc[0]
X_test['Cabin'].fillna(most_frequent_title, inplace=True)

most_frequent = X_test['Fare'].mode().iloc[0]
X_test['Fare'].fillna(most_frequent, inplace=True)

most_frequent_TICKET = X_train['Embarked'].mode().iloc[0]
X_train['Embarked'].fillna(most_frequent_title, inplace=True)

most_frequent_title = X_test['Embarked'].mode().iloc[0]
X_test['Embarked'].fillna(most_frequent_title, inplace=True)

```
